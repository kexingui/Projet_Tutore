### trouve numéro de locuteur

```python
import csv
import re


csv_file_path = '/Users/kexingui/Downloads/01augmented_dataset.csv'

#numéro de locuteur
pattern = re.compile(r'ESLO2_ENT_\d+_(\w+)_\d+\.txt')


codes = []


with open(csv_file_path, mode='r', encoding='utf-8') as file:
    reader = csv.DictReader(file)
    for row in reader:
        filename = row['filename']
        match = pattern.search(filename)
        if match:
            code = match.group(1)  
            codes.append(code)

for code in codes:
    print(code)
```

### récupérer le html de chaque locuteur


```python
import os
from bs4 import BeautifulSoup


folder_path = '/Users/kexingui/Downloads/HTMLs'


code_links = {}

#  parcourire HTML
for filename in os.listdir(folder_path):
    if filename.endswith('.html'):
        file_path = os.path.join(folder_path, filename)
        
      
        with open(file_path, 'r', encoding='utf-8') as file:
            soup = BeautifulSoup(file, 'html.parser')
            
           
            links = soup.find_all('a', class_='lien')
            
            for link in links:
                href = link.get('href')  # 
                text = link.get_text(strip=True)  
                
                if text in codes:
                    
                    code_links[text] = href


for code, link in code_links.items():
    print(f'{code}: {link if link else "No link found"}')
```

### scrach infos

```python
import csv
import requests
from bs4 import BeautifulSoup

# CSV head
csv_headers = ['Code', 'Niveau d\'études', 'Catégorie Professionnelle (INSEE)', 'Profession en termes propres', 'Commentaire']


csv_filename = 'extracted_info2.csv'

with open(csv_filename, 'w', newline='', encoding='utf-8') as csv_file:
    writer = csv.writer(csv_file)
    writer.writerow(csv_headers)
    
    # parcourire code et link
    for code, link in code_links.items():
        response = requests.get(link)
        soup = BeautifulSoup(response.content, 'html.parser')
        
        # récupérer info
        niveau_etudes = soup.find('th', text='Niveau d\'études:').find_next_sibling('td').text.strip() if soup.find('th', text='Niveau d\'études:') else 'N/A'
        categorie_pro = soup.find('th', text='Catégorie Professionnelle (INSEE):').find_next_sibling('td').text.strip() if soup.find('th', text='Catégorie Professionnelle (INSEE):') else 'N/A'
        profession = soup.find('th', text='Profession en termes propres:').find_next_sibling('td').text.strip() if soup.find('th', text='Profession en termes propres:') else 'N/A'
        commentaire = soup.find('th', text='Commentaire:').find_next_sibling('td').text.strip() if soup.find('th', text='Commentaire:') else 'N/A'
        
        
        writer.writerow([code, niveau_etudes, categorie_pro, profession, commentaire])
```
